```python
def funding(x):
    if x<10000000:
        return 'Low Funding'
    if x>=10000000 and x<=1000000000:
        return 'Medium Funding'
    if x>1000000000:
        return 'High Funding'
```

```python
dfn['Funding_type']=dfn['funding_total_usd'].apply(funding)
```

```python
dfn
```

```python
#creating feature company_type
```

```python
def company_type(x):
    if x<=1942 and x>=1906:
        return 'old-age company'
    if x>1942 and x<=1978:
        return 'mid-age company'
    if x>1978:
        return 'young-age company'
```

```python
dfn['company_type']=df['founded_at'].apply(company_type)
```
